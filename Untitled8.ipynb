```python
from pypdf import PdfReader


# creating a pdf reader object

reader = PdfReader('ESSAY IN HINDI SSC Forum.pdf')


# printing number of pages in pdf file

print(len(reader.pages))
```

```python
# getting a specific page from the pdf file

page = reader.pages[1]
```

```python
# extracting text from page

text = page.extract_text()
```

```python
print (text)
```

```python
from transformers import MT5ForConditionalGeneration, MT5Tokenizer

tokenizer = MT5Tokenizer.from_pretrained("csebuetnlp/mT5_multilingual_XLSum")

model =  MT5ForConditionalGeneration.from_pretrained("csebuetnlp/mT5_multilingual_XLSum")

tokens_input = tokenizer.encode("summarize: "+text, return_tensors='pt', max_length=4001, truncation=True)
summary_ids = model.generate(tokens_input, min_length=100, max_length=200)
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
print(summary)
```
